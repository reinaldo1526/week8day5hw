# Nba Regular Season 2018-19 Data Challenge

Your task will be to take the dataset given, and create an analysis answering the following 10 questions. This project will again test your knowledge of pandas in order to find the answers needed given the data you are presented with.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
nba = pd.read_csv('files/nbastats2018-2019.csv')
nba.head()
```

# What was the average age of player in the league?

```python
nba['Age'].mean()
#25 is the average age
```

# What player scored the most points?

```python
most_points = nba.groupby(['Name'])[['Points']].max().sort_values(by='Points')
most_points
#James Harden 
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
most_Blocks = nba.groupby(['Name'])[['Blocks']].max().sort_values(by='Blocks')
most_Blocks
#Myles Turner had the most Blocks he plays center
```

# Based on the regular season, who had the best chance to win a title given their win percentage?

```python
most_win = nba.groupby(['Team'])[['WS']].sum().sort_values(by='WS')
most_win

#Milwaukee Bucks has the most win share 63.0
```

# What player had the best 3-pt percentage? 

```python
bst3 = nba.fillna('N/A')
bst3[bst3['FG3%'] != 'N/A'][['Name','FG3%']].sort_values('FG3%')
# Jordan Siber, Trevon Duval, Scott Machado, Eric Moreland
```

# Who played the most minutes during the season

```python
#nba.sort_values('MP')
most_min = nba.groupby(['Name'])[['MP']].sum().sort_values(by='MP')
most_min

#Bradley Beal,Paul George
```

# What player given their player effiecency rating was the clutchest during the season?

```python
rating = nba.groupby(['Name'])[['PER']].sum().sort_values(by='PER')
rating
#Zhou Qi had an efficiency rating of 80.4
```

# What team had the youngest roster?

```python
youngest = nba.groupby(['Team'])[['Age']].mean()
youngest
#Chicago Bulls
```

# Who is the highest paid player during the seasion?

```python
salary = nba.groupby(['Name'])[['Salary']].sum().sort_values(by='Salary')
salary
#Dwight Powell
```

# At the end of a game, who WOULDN'T you want on the Free Throw Line?

```python
nba[['Name','FT%']].sort_values('FT%')

#Terrence Jones and Haywood Highsmith
```
